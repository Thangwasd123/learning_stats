# Generating random Distribution

```python

rng = np.random.default_rng(seed=20260429)
true_p = 0.3         
n = 200             
data = rng.binomial(n=1, p=true_p, size=n)
print(data)
```

```python
len(data)
```

```python
uniq = np.unique_counts(data)
print(uniq.values)
print(uniq.counts)
probability_vector = uniq.counts/len(data)
print(probability_vector)
```

# Solving with MLE

\begin{equation}
\LARGE L(p) = \prod_{i=1}^{n} p^{x_i} \cdot (1-p)^{1-x_i}
\end{equation}


```python
def log_likelihood(vector):
    result = np.unique_counts(vector)
    counts = dict(zip(result.values, result.counts))  # {0: n0, 1: n1}
    total = 0.0
    for val in probability_vector:
        p = probability_vector[val]
        n = counts.get(val, 0)
        total += n * np.log(p) + (len(vector) - n) * np.log(1 - p)
    return total


print(log_likelihood(data))
```
