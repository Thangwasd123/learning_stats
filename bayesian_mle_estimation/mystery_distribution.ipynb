# Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from scipy.stats import gaussian_kde

plt.style.use("bmh")
```

# Generating random Distribution - Binomial Distribution

```python

rng = np.random.default_rng(seed=20260429)
true_p = 0.3         
n = 200             
data = rng.binomial(n=1, p=true_p, size=n)
print(data)
```

# Solving with MLE

\begin{equation}
\LARGE L(p) = \prod_{i=1}^{n} p^{x_i} \cdot (1-p)^{1-x_i}
\end{equation}

\begin{equation}
\frac{d}{dp} \left[ n_1 \log(p) + n_0 \log(1-p) \right] = \frac{n_1}{p} - \frac{n_0}{1-p} = 0
\end{equation}



```python
def log_likelihood(p, data):
    n1 = data.sum()          # count of 1s
    n0 = len(data) - n1      # count of 0s
    return n1 * np.log(p) + n0 * np.log(1 - p)

# Evaluate over grid of p values
p_values = np.linspace(0.01, 0.99, 100)
log_likelihoods = [log_likelihood(p, data) for p in p_values]  # no outer loop needed

p_mle = p_values[np.argmax(log_likelihoods)]
likelihoods_vector = np.array(log_likelihoods)

print(p_mle)
print(p_values)
print(likelihoods_vector)
```

```python
fig = px.line(x=p_values, y=likelihoods_vector, title="Probability of distribution P-values")

fig.add_scatter(
    x=[p_mle], 
    y=[max(likelihoods_vector)], 
    mode="markers",
    marker=dict(color="red", size=10),
    name=f"MLE = {p_mle:.3f}"
)

fig.update_layout(
    xaxis_title="p-values", yaxis_title="Corresponding Log-Likelihood"
)
fig.show()
```

# Gaussian Distribution

```python
rng = np.random.default_rng(seed=42)
true_mu    = 5.0
true_sigma = 2.0

data = rng.normal(loc=true_mu, scale=true_sigma, size=100)
```

```python
kde = gaussian_kde(data)
x_range = np.linspace(data.min(), data.max(), 200)

fig = px.histogram(x=data, histnorm="probability density")  
fig.add_scatter(x=x_range, y=kde(x_range), mode="lines", name="KDE")
fig.show()
```
