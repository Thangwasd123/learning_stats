# Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from scipy.stats import gaussian_kde
from plotly.subplots import make_subplots
from scipy.stats import norm
import matplotlib.pyplot as plt

plt.style.use("bmh")
```

# Generating random Distribution - Gaussian Distribution

```python
rng = np.random.default_rng(seed=20260502)
true_mu    = 5.0
true_sigma = 2.0
data = rng.normal(loc=true_mu, scale=true_sigma, size=100)
```

```python
kde = gaussian_kde(data)
x_range = np.linspace(data.min(), data.max(), 200)

fig = px.histogram(x=data, histnorm="probability density")  
fig.add_scatter(x=x_range, y=kde(x_range), mode="lines", name="KDE")
fig.show()
```

```python

mu_grid = np.linspace(-5, 15, 500)    # x axis — range of plausible μ values

# your prior belief: μ ~ N(μ₀, σ₀²)
mu_0    = 0.0    # "I think the mean is around 0"
sigma_0 = 5   # "but I'm very uncertain"

prior = norm.pdf(mu_grid, loc=mu_0, scale=sigma_0)
```

```python

plt.plot(mu_grid, prior)        # x = candidate μ values
                                # y = probability density
plt.xlabel("μ")
plt.ylabel("density")
plt.title("Prior distribution")
plt.show()
```
